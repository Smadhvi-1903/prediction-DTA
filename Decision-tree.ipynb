# importing libraries

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn import tree
import sklearn.datasets as datasets
import pandas as pd
```

# importing dataset

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

# Define the Decision Tree Algorithm

```python
models = DecisionTreeClassifier(random_state=1234)
model = models.fit(X,y)
```

# lets visualize Textually the Decision Tree Algorithm

```python
text_representation = tree.export_text(models)
print(text_representation)
```

# visualize the Decision Tree to understand it better

```python
fig = plt.figure(figsize = (25,20))
_ = tree.plot_tree(models, feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```
